```python

import akshare as ak

cons = ak.index_component_sw(symbol="801010")

# 获取成分股收盘价
close_prices = []
for code in cons['证券代码']:
    # df = ts.get_hist_data(code)
    df = ak.stock_zh_a_hist(symbol=code, period="daily", start_date="20220301", end_date='202303024', adjust="")
    close_price = df['收盘']
    close_prices.append(close_price)

# 获取成分股权重
weights = cons['最新权重'].tolist()

# 计算上证指数
index_value = sum([close_prices[i] * weights[i] for i in range(len(close_prices))]) / 100

print('上证指数：', index_value)
```

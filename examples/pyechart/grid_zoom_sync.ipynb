```python
import random
from pyecharts import options as opts
from pyecharts.charts import Line, Grid, Bar

x_data = [str(i) for i in list(range(1990, 2020))]

y_data_1 = [random.randint(0, 100) for _ in x_data]
y_data_2 = [random.randint(0, 100) for _ in x_data]


line1 = (
    Line()
    .add_xaxis(x_data)
    .add_yaxis(series_name='图例1', y_axis=y_data_1,xaxis_index=0, yaxis_index=0)
    # .set_global_opts(datazoom_opts=[opts.DataZoomOpts(range_start=50, range_end=80,xaxis_index=0,pos_top="80%"),opts.DataZoomOpts(range_start=50, range_end=80,xaxis_index=1,pos_top="90%")])
    .set_global_opts(datazoom_opts=[
        opts.DataZoomOpts(is_show=False, start_value=1995, end_value=2005,
                          xaxis_index=[0, 0], pos_top="90%"),
        opts.DataZoomOpts(is_show=True, start_value=1995, end_value=2005,
                          xaxis_index=[0, 1], pos_top="90%"),
        opts.DataZoomOpts(is_show=False, start_value=1995, end_value=2005,
                          xaxis_index=[0, 2], pos_top="90%")
    ])
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
)


line2 = (
    Line()
    .add_xaxis(xaxis_data=x_data)
    .add_yaxis(series_name='图例2', y_axis=y_data_2,xaxis_index=1, yaxis_index=1)
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))

)

bar1 = (
    Bar()
    .add_xaxis(xaxis_data=x_data)
    .add_yaxis(series_name='图例3', y_axis=y_data_2,xaxis_index=2, yaxis_index=2)
    .set_series_opts(label_opts=opts.LabelOpts(is_show=False))
)

grid = Grid()
grid.add(line1, grid_opts=opts.GridOpts(pos_top="10%", pos_bottom="60%"))
grid.add(line2, grid_opts=opts.GridOpts(pos_top="50%", pos_bottom="30%"))
grid.add(bar1, grid_opts=opts.GridOpts(pos_top="80%", pos_bottom="10%"))


# Render the chart
grid.render_notebook()
```

```python
from czsc import CZSC
from czsc.sensors.stocks import TsDataCache
```

```python
#TsDataCache：tushare 临时性数据缓存，单次缓存，多次使用，不做增量更新，适用于研究场景
dc = TsDataCache(data_path=r"C:\ts_data", sdt='2000-01-01', edt='2022-02-18')
dc.pro.__token='60348c9924c35ceb20e8fa05df697f7304537b511b45def253d713c3'
# pro_bar tushare 通用行情接口
bars = dc.pro_bar(ts_code='600171.SH', start_date='20150101', end_date="20220427", freq='D')

czsc_analysis = CZSC(bars)
```

```python
# 在jupyter中打开czsc分析
czsc_analysis.to_echarts().render_notebook()
```

```python
# 在浏览器中打开czsc分析
czsc_analysis.open_in_browser()
```
